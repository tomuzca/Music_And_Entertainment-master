```python
import matplotlib.pyplot as plt
columns_to_plot = ["danceability_%", "valence_%", "acousticness_%", "instrumentalness_%", "liveness_%", "speechiness_%"]

for column in columns_to_plot:
    plt.figure(figsize=(15, 6))
    plt.bar(spotify_2023["released_year"], spotify_2023[column], alpha=0.5)
    plt.title(f"{column} vs released_year")
    plt.xlabel("Released Year")
    plt.ylabel(column)
    plt.grid(True)
    plt.show()
```

```python
songs_per_year = spotify_2023['released_year'].value_counts().sort_index()

# Crear el grÃ¡fico de barras
plt.figure(figsize=(10, 6))
songs_per_year.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Canciones por AÃ±o')
plt.xlabel('AÃ±o de Lanzamiento')
plt.ylabel('Cantidad de Canciones')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
